```python
def cohort(tenure):
    if tenure<13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure < 49:
        return '24-48 Months'
    else:
        return 'Over 48 Months'
```

```python
df['tenure']
```

```python
df['tenure'].apply(cohort)
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
df[['Tenure Cohort', 'tenure']]
```

## Part 3 

```python
X = df.drop(['customerID','Churn'],axis=1)
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=6)
```

```python
dt.fit(X_train,y_train)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix,classification_report
```

```python
pred = dt.predict(X_test)
```

```python
cm = confusion_matrix(y_test,pred)
```

```python
cm
```

```python
print(classification_report(y_test,pred))
```

```python
CM = ConfusionMatrixDisplay.from_predictions(y_test,pred)
```

```python
dt.feature_importances_
```

```python
imp_feats = pd.DataFrame(data = dt.feature_importances_,
                         index = X.columns,
                         columns = ['Feat IMP'])
```

```python
imp_feats
```

```python
imp_feats.sort_values('Feat IMP')
```

```python
imp_feats = imp_feats.sort_values('Feat IMP')
```
