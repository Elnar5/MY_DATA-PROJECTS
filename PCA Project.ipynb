## Scaling Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_pixels = scaler.fit_transform(pixels)
```

```python
scaled_pixels
```

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pca_model.fit(scaled_pixels)
```

```python
labels = pca_model.transform(scaled_pixels)
```

```python
labels
```

```python
pixels.columns
```

```python
data = pca_model.components_
```

```python
data
```

```python
data_frame = pd.DataFrame(data = data,index=['PC1','PC2'],columns = pixels.columns)
```

```python
data_frame
```

```python
pca_model.explained_variance_ratio_
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
plt.figure(figsize = (10,10),dpi=150)
sns.scatterplot(y= labels[:,0], x = labels[:,1],data = pd.DataFrame(scaled_pixels),hue= digits['number_label'])
```

```python
plt.figure(figsize = (10,10),dpi=150)
sns.scatterplot(x= labels[:,0], y = labels[:,1],data = digits,hue= 'number_label',palette='tab10')
```
